```python
import pandas as pd
import numpy as np
```

```python
cars = pd.read_excel('CARS_English.xlsx', sheet_name="Sheet1", usecols=range(10))
cars
```

```python
# change gender to binary
cars['gender'] = cars['gender'].map({'female':0, 'male': 1})

# change gender column name to is_male
cars.rename(columns={'gender':'is_male'}, inplace=True)

cars.head()
```

```python
# remove rows where month_payment is "-"
cars = cars[cars.month_payment != '-']
```

```python
# get ages where age isnt #
ages = cars[cars.age != '#'].age

# strip whitespace from ages
ages = ages.astype(str).str.strip()

# remove - values from ages
ages = ages.str.replace('-', '')

# convert ages to int
ages = ages.astype(int)

# apply to dataframe
cars.age = ages

cars.head()
```

```python
# drop all rows where month payment is 0
cars = cars[cars.month_payment != 0]

# remove rows where month_payment is "-"
cars = cars[cars.month_payment != '-']

# change month_payment to float
cars.loc[:, 'month_payment'] = cars.month_payment.astype(float)
```

```python
# drop payment_duration column
cars = cars.drop(columns=['payment_duration'])
```

```python
cars['payment_duration'] = (cars.car_price - cars.scum_mimon) / cars.month_payment
cars.head()
```

```python
# move payment_duration to 8th column
cars.insert(7, 'payment_duration', cars.pop('payment_duration'))

cars.head()
```

```python
cars.to_excel('Cars_English_v2.xlsx', index=False)
```
