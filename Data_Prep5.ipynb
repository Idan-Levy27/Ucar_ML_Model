```python
import numpy as np
import pandas as pd
```

```python
cars = pd.read_excel('CARS_English_v4.xlsx', sheet_name="Sheet1", usecols=range(12))
cars
```

```python
# fix payment duration

cars['payment_duration'] = (cars['scum_mimon'] / cars['month_payment']).round(2)
cars
```

```python
# remove row 5899 (0 scum mimon)
cars = cars.drop(5899)
```

```python
# amount of cars with more than 20 years of payment duration
len(cars[cars['payment_duration'] > 240])
```

```python
# drop all payment_duration above 240 months (20 years)
cars = cars.drop(cars[cars['payment_duration'] > 240].index)
cars
```

```python
# change scum_mimon column name to funding_amount
cars.rename(columns={'scum_mimon': 'funding_amount'}, inplace=True)
cars
```

```python
cars['funding_percent'] = ((cars['funding_amount'] / cars['car_price'])*100).round(2)
cars
```

```python
cars.insert(7, 'fundingPercent', cars.funding_percent)
cars.drop(['funding_percent'], axis=1, inplace=True)
cars
```

```python
# change iwWorkCar to 0 if model the Outlander
# (fixing mistake)
cars.loc[cars['model'] == 'Outlander', 'isWorkCar'] = 0
```

```python
cars.to_excel('Cars_English_v5.xlsx', index=False)
```
