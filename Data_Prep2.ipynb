```python
import pandas as pd
import numpy as np
```

```python
cars = pd.read_excel('CARS_English_v2.xlsx', sheet_name="Sheet1", usecols=range(10))
cars
```

We have 1402 rows where the age value is incorrect:

```python
# show all cars with age lower than 18 or bigger than 90
len(cars[(cars['age'] < 18) | (cars['age'] > 90)])
```

```python
# store all the indexes of the rows that have incorrect age values
wrong_age_id = cars[(cars['age'] < 18) | (cars['age'] > 90)].index
wrong_age_id
```

```python
# show all rows with indexes from wrong_age_id
cars.loc[wrong_age_id]
```

```python
len(cars[cars['model'] == '-'])
```

```python
# drop all rows where model is "-" in them
cars.drop(cars[cars['model'] == '-'].index, inplace=True)
cars.loc[(cars['age'] < 18) | (cars['age'] > 90), 'age'] = np.NaN
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse=False)
encoded_brand_model = encoder.fit_transform(np.asarray(cars[['brand', 'model']]))
```

```python
# concat encoded_brand_model to cars
cars_encoded = pd.concat([cars, pd.DataFrame(encoded_brand_model)], axis=1)

# drop brand and model 
cars_encoded = cars_encoded.drop(['brand', 'model', 'num_id'], axis=1)
cars_encoded.head()
```

```python
cars_encoded.loc[wrong_age_id]
```

```python
# find all nulls in cars encoded
cars_encoded.iloc[:, :7].isna().sum()
```

```python
from sklearn.impute import KNNImputer

imputer = KNNImputer()
imputed_age = imputer.fit_transform(cars_encoded)
imputed_cars = pd.DataFrame(imputed_age, columns=cars_encoded.columns)

#cars_encoded['age'] = imputed_age
#cars_encoded.head()
```

```python
imputed_cars.loc[wrong_age_id]
```

```python
cars['age'] = imputed_cars['age']
cars.head()
```

```python
cars.to_excel('Cars_English_v3.xlsx', index=False)
```
