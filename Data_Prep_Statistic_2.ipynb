```python
from scipy.stats import shapiro

stat, p = shapiro(cars.age)
print('Statistics=%.3f, p=%.3f' % (stat, p))

alpha = 0.05
if p > alpha:
    print('Sample looks Gaussian (fail to reject H0)')
else:
    print('Sample does not look Gaussian (reject H0)')
```

```python
fes
femaleAgreePrecent=sum(fes)/(len(fes))
maleAgreePrecent=sum(mes)/(len(mes))
print("Female accept rate:  %.2f , Male accept rate: %.2f" % (femaleAgreePrecent,maleAgreePrecent))
```

```python
from scipy.stats import chi2_contingency
g, p, dof, expctd = chi2_contingency([[sum(fes), sum(mes)],[len(fes)-sum(fes), len(mes)-sum(mes)]])
g, p, dof, expctd

print('stat=%.2f, p=%.2f' % (g, p))
if p > 0.05:
  print('Probably the same distribution')
else:
  print('Probably different distributions')
  if(femaleAgreePrecent<maleAgreePrecent):
    print('Men are more likely to accept an offer compared to women')
  else:
    print('Women are more likely to accept an offer compared to men')
```

```python
feDuration=np.asarray(females['payment_duration'])
meDuration=np.asarray(males['payment_duration'])
```

```python
femaleAvgPaymentDur=feDuration.mean()
maleAvgPaymentDur=meDuration.mean()
print("Female avg payment duration:  %.2f , Male avg payment duration: %.2f" % (femaleAvgPaymentDur,maleAvgPaymentDur))
females['payment_duration'].describe()
```

```python
males['payment_duration'].describe()
```

```python
from scipy.stats import ks_2samp
stat, p = ks_2samp(feDuration, meDuration)

print('stat=%.2f, p=%.4f' % (stat, p))
if p > 0.05:
  print('Probably the same distribution')
else:
  print('Probably different distributions')
  if(meDuration.mean()>feDuration.mean()):
    print('Men payment duration is longer than women payment duration')
  else:
    print('Women payment duration is longer than men payment duration')
```

```python
stat, p = stats.mannwhitneyu(feDuration, meDuration)

print('stat=%.2f, p=%.4f' % (stat, p))
if p > 0.05:
  print('Probably the same distribution')
else:
  print('Probably different distributions')
  if(meDuration.mean()>feDuration.mean()):
    print('Men payment duration is longer than women payment duration')
  else:
    print('Women payment duration is longer than men payment duration')
```

We can understand by the p-value that the payment duration isn't different for men and women

Average funding percent for work cars / private cars

```python
workCars = cars[cars['isWorkCar'] == 1]
privateCars = cars[cars['isWorkCar'] == 0]

workCarsFundingPercent = np.asarray(workCars['fundingPercent'])
privateCarsFundingPercent = np.asarray(privateCars['fundingPercent'])

workCarsAvgFundingPercent = workCarsFundingPercent.mean()
privateCarsAvgFundingPercent = privateCarsFundingPercent.mean()

print("Work Car avg funding percent: %.2f , Private Car avg funding percent: %.2f" % (workCarsAvgFundingPercent, privateCarsAvgFundingPercent))
```

```python
stat, p = stats.mannwhitneyu(workCarsFundingPercent, privateCarsFundingPercent)

print('stat=%.2f, p=%.4f' % (stat, p))
if p > 0.05:
  print('Probably the same distribution')
else:
  print('Probably different distributions')
  if(workCarsAvgFundingPercent > privateCarsAvgFundingPercent):
    print('Work cars usually get more funding than private cars')
  else:
    print('Private cars usually get more funding than work cars')
```

Compare funding percent of sold/not sold offers

```python
fundingPercentSold = np.asarray(sold['fundingPercent'])
fundingPercentNotSold = np.asarray(notSold['fundingPercent'])

fundingAvgPercentSold = fundingPercentSold.mean()
fundingAvgPercentNotSold = fundingPercentNotSold.mean()

print("Sold cars avg funding percent: %.2f, Not sold cars avg funding percent: %.2f" % (fundingAvgPercentSold, fundingAvgPercentNotSold))
```

```python
stat, p = stats.mannwhitneyu(fundingPercentSold, fundingPercentNotSold)

print('stat=%.2f, p=%.4f' % (stat, p))
if p > 0.05:
  print('Probably the same distribution')
else:
  print('Probably different distributions')
  if(fundingAvgPercentSold > fundingAvgPercentNotSold):
    print('Offers that were accepted had higher funding than offers that were not accepted')
  else:
    print('Offers that were not accepted had higher funding than offers that were accepted')
```

Compare seats amount of sold/not sold offers

```python
seatsAmountSold = np.asarray(sold['seatsAmount'])
seatsAmountNotSold = np.asarray(notSold['seatsAmount'])

seatsAvgAmountSold = seatsAmountSold.mean()
seatsAvgAmountNotSold = seatsAmountNotSold.mean()

print("Sold cars avg seats amount: %.2f, Not sold cars avg seats amount: %.2f" % (seatsAvgAmountSold, seatsAvgAmountNotSold))
```

```python
stat, p = stats.mannwhitneyu(seatsAmountSold, seatsAmountNotSold)

print('stat=%.2f, p=%.4f' % (stat, p))
if p > 0.05:
  print('Probably the same distribution')
else:
  print('Probably different distributions')
  if(fundingAvgPercentSold > fundingAvgPercentNotSold):
    print('Offers that were accepted had more seats than offers that were not accepted')
  else:
    print('Offers that were not accepted more seats than offers that were accepted')
```

```python
# find all correlated values

corr_matrix = df_corr.abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))

to_drop = [(r, c) for r in upper.columns for c in upper.columns if upper.loc[r, c] >= 0.8 and upper.loc[r, c] != 1]
print(to_drop)
```
